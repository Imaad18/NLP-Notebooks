# **Key NLP Techniques and WorkFlow:**

* NLP pipelines typically involve three main stages:

**A. Text Processing:**

* **Tokeniztion:** Splits Text into words , subwords, or sentences
* **Example:** "This is the Era of AI",  into "This","is","the" , "Era", "of" , "AI".

* **Stop Words Removal:** Filters out common, low-value words (eg., "the","and").

* **Stemming/Lemmatization:**  Reduces words to their root form for consistency (eg., "running"~"run").

* **Cleaning:** Handles case normalization, punctuation, and special characters.



```python
# An Example in Python how these work

import nltk
import re   # re = regular expression to find search pattern
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python
# Download Required Data Files
nltk.download("punkt")
nltk.download("stopwords")
nltk.download("wordnet")
```

```python
# Sample Text
text = "The cats are running fatser than the dogs, and they haven't stopped!"

# ========== 1. Cleaning : Lowercase and remove special characters ============
cleaned_text = re.sub(r'[^a-zA-Z\s]', '', text.lower())
print("Cleaned Text:", cleaned_text)

# ========= 2. Tokenization: Split into words =======================
tokens = word_tokenize(cleaned_text)
print("Tokens:", tokens)

# =========== 3. Stop Word Removal =========================
stop_words = set(stopwords.words("english"))
filtered_tokens = [word for word in tokens if word not in stop_words]
print("After Stop Word Removal:", filtered_tokens)

# ============ 4. Stemming ======================
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_tokens]
print("Stemmed Words:", stemmed_words)

# ========  5. Lemmatization ======================
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_tokens]
print("Lemmatized Words:", lemmatized_words)
```

**B. Feature Extraction:**

* **Bag-of-Words(BoW):** Represents text as word frequency counts.

* **TF-IDF:** Weighs words by their importance across documents.

* **Word Embeddings:** Dense vector representations (Word2Vec, GloVe) that capture semantic relationships between words.

* **Contextual Embeddings:** Modeen models like BERT generate embeddings that chnage based on word context, addressing ploysemy (eg. "bank" as a financial institution vs. river edge)

* **Machine Learning & Deep Learning Models:**

 * **1.Classical Models:** Naive Bayes, SVM, and Logistic Regression for basic tasks.

 * **2.Neural Architectures:** RNNs, LSTMs, CNNs and especially transformers for advanced, contextaware tasks.
